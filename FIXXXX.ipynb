```python
#Return Daily
return_daily = data_normalize.pct_change(1).mul(100).dropna().copy()
return_daily
```

```python
#Return Annualize %
return_ann = return_daily.mean()*252
return_ann
```

```python
# ROR PORTO 
weight = np.array([0.33,0.33,0.33,0.33,0.33,0.33,0.33,0.33,0.33,0.33,0.33,0.33,0.33,0.33,0.33,0.33,0.33,0.33,0.33,0.33,
                   0.33,0.33,0.33,0.33,0.33,0.33,0.33,0.33,0.33,0.33,0.33,0.33,0.33,0.33,0.33,0.33,0.33,0.33,0.33,0.33,
                   0.33,0.33,0.33,0.33,0.33])
np.dot(return_ann,weight)
```

```python
pfoloio_1 = str(round(np.dot(return_ann,weight),2)) + ' %'
pfoloio_1
```

```python
np.sum(weight * return_daily.mean()*252)
```

```python
weight * return_daily.mean()*252 #Expected Return
```

```python
# STD Porto
pfolio_1_std = np.sqrt((np.dot(weight.T, np.dot(return_daily.cov()*252,weight))))
pfolio_1_std
```

```python
str(round(pfolio_1_std,2)) + ' %'
```
