```python
# libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
#original image
image = cv2.imread('/content/jaehyun ganteng.jpg',1)
cv2_imshow(image)
```

```python
image.shape
```

```python
def rgb2gray(rgb):

    r, g, b = rgb[:,:,0], rgb[:,:,1], rgb[:,:,2]
    gray = 0.2989 * r + 0.5870 * g + 0.1140 * b

    return gray
gray=rgb2gray(image)
cv2_imshow(gray)
```
